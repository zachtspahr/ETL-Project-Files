```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
pre_condition = pd.read_csv("pre-existing-conditions-by-congressional-district -number with pre-ex by district.csv")

pre_condition.head()
```

```python
pre_condition ["District"] = pre_condition ['District'].str.replace(')', '')
pre_condition ["District"] = pre_condition ['District'].str.replace('-', '')
pre_condition ["Age_0_to_17_with_Pre-existing_Condition"] = pre_condition ["Age_0_to_17_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Age_18_to_24_with_Pre-existing_Condition"] = pre_condition ["Age_18_to_24_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Age_25_to_34_with_Pre-existing_Condition"] = pre_condition ["Age_25_to_34_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Age_35_to_44_with_Pre-existing_Condition"] = pre_condition ["Age_35_to_44_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Age_45_to_54_with_Pre-existing_Condition"] = pre_condition ["Age_45_to_54_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Age_55_to_64_with_Pre-existing_Condition"] = pre_condition ["Age_55_to_64_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Nonelderly_with_Pre_existing_Condition"] = pre_condition ["Nonelderly_with_Pre-existing_Condition"].str.replace(',', '')
pre_condition ["Percent_of_Nonelderly_with_Pre-existing_Condition"] = pre_condition ["Percent_of_Nonelderly_with_Pre-existing_Condition"].str.replace('%', '')

pre_condition = pre_condition.drop(columns = ["Representative (District)"])
pre_condition.head(8)
```

```python
pre_condition_with_index = pre_condition.set_index("District")
pre_condition_with_index ["Percent_of_Nonelderly_with_Pre-existing_Condition"] = pre_condition_with_index ["Percent_of_Nonelderly_with_Pre-existing_Condition"].astype(float)
pre_condition_with_index ["Age_0_to_17_with_Pre-existing_Condition"] = pre_condition_with_index ["Age_0_to_17_with_Pre-existing_Condition"].astype(int)
pre_condition_with_index ["Age_18_to_24_with_Pre-existing_Condition"] = pre_condition_with_index ["Age_18_to_24_with_Pre-existing_Condition"].astype(int)
pre_condition_with_index ["Age_25_to_34_with_Pre-existing_Condition"] = pre_condition_with_index ["Age_25_to_34_with_Pre-existing_Condition"].astype(int)
pre_condition_with_index ["Age_35_to_44_with_Pre-existing_Condition"] = pre_condition_with_index ["Age_35_to_44_with_Pre-existing_Condition"].astype(int)
pre_condition_with_index ["Age_45_to_54_with_Pre-existing_Condition"] = pre_condition_with_index ["Age_45_to_54_with_Pre-existing_Condition"].astype(int)
pre_condition_with_index ["Age_55_to_64_with_Pre-existing_Condition"] = pre_condition_with_index ["Age_55_to_64_with_Pre-existing_Condition"].astype(int)
pre_condition_with_index ["Nonelderly_with_Pre_existing_Condition"] = pre_condition_with_index ["Nonelderly_with_Pre_existing_Condition"].astype(int)
pre_condition = pre_condition_with_index.reset_index()
pre_condition.head()
```

```python
cd_data  = pd.read_csv("politicaldata.csv")
cd_data_cleaned_excel = pd.read_csv("politicaldata_cleaned_on_excel.csv")

cd_data_cleaned_excel.head(10)
```

```python
merge_table = pd.merge(pre_condition, cd_data, on="District")
merge_table2 = pd.merge(pre_condition, cd_data_cleaned_excel, on="District")
```

```python
merge_table ["District"] = merge_table2 ['CD']
merge_table ["Representative"] = merge_table2 ['2018_Winner']
merge_table.iloc [:,0:18]
# This made this merge table updated with better data from the congressional districts as far as who represents 
# each district today. 
```

```python
merge_table2 ["Representative"] = merge_table2 ['2018_Winner']
merge_table2 ["District"] = merge_table2 ['CD']
merge_table2 = merge_table2.drop(columns = ["2018_Cook_PVI_Score_intg",'2018_Winner',"CD","State_y"])
# Note I' leaving in State and CD # columns because that is the best way to sort this data correctly if needed

merge_table2.iloc [:,0:18]
```

```python
merge_table_with_index = merge_table.set_index("District")
merge_table2_with_index = merge_table2.set_index("District")
```
